### 2.3.1 Задача. Найдитн евклидово расстояние между двумя Series (точками) a и b, не используя встроенную формулу.

```python
import numpy as np
a = np.array((3, 0, 3))
b = np.array((3, 3, 3))

square = np.square(a - b)
sum_square = np.sum(square)

distance = np.sqrt(sum_square)
print(distance)
```
